```python

%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
import bqplot.pyplot as bqpyplt
```

The quadratic equation:

>$y= f(x) = ax^2+bx+c$

>focus $= (0,\frac{1}{4}a)$

>axis of symetry $ = \frac{-b}{2a}$

>exponenial_func $ = ab^x$

```python
def f(x):
  return a*(x**2)+b*x+c

a=1
b=1
c=2000
x_cords = range(-100,100)
y_cords = [f(x) for x in x_cords]
plt.plot(x_cords, y_cords, color='green')
plt.scatter([0], [1/4*a], marker='x', color='brown' )

a=1
b=1
c=0
x_cords = range(-100,100)
y_cords = [f(x) for x in x_cords]
plt.plot(x_cords, y_cords, color='red')
plt.scatter([0], [1/4*a], marker='x', color='blue' )


plt.show()
```

```python
fig = bqpyplt.figure(title='First Example')
bqpyplt.plot(x_cords,y_cords)
fig
```

```python


def f(m, b):
    plt.figure(2)
    x = np.linspace(-10, 10, num=1000)
    plt.plot(x, m * x + b)
    plt.ylim(-5, 5)
    plt.show()

interactive_plot = interactive(f, m=(-2.0, 2.0), b=(-3, 3, 0.5))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```
